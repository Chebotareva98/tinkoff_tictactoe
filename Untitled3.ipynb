```python
def promo_count(data):
  days_count = 1
  promo_count = {}
  for i in range(len(data) - 1):
    current = data.iloc[i]
    next = data.iloc[i + 1]
    if (next[1] - current[1]) == 1 and current[0] == next[0]:
      days_count += 1
    else:
      promo_count[int(current[0])] = (promo_count[int(current[0])] if int(current[0]) in promo_count else 0) + 1
      days_count = 1
  return promo_count  
```

```python

print(promo_count(promo_first))
```
